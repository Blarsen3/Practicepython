# Question 6: Average of numbers.
Assume a file containing a series of integers is named numbers.txt and exists on the computer’s disk. Write a program that calculates the average of all the numbers stored in the file

## First you want to Define the filename: The file containing the integers is named numbers.txt. Initialize variables: Variables total and count are initialized to store the sum of the numbers and the count of the numbers, respectively.Open the file in read mode: The open function with the with statement opens the file in read mode ("r").Read all lines: The readlines method reads all lines from the file into a list.

Lest do that below 

```python
# Define the filename
filename = "numbers.txt"

# Open the file in write mode
with open(filename, "w") as file:
    # Write your name to the file
    file.write("5")



# Define the filename
filename = "numbers.txt"

# Initialize variables to store the sum of numbers and the count of numbers
total = 0
count = 0

# Open the file in read mode
with open(filename, "r") as file:
    # Read all lines from the file
    lines = file.readlines()
    
    # Process each line
    for line in lines:
        # Convert the line to an integer and add it to the total
        number = int(line.strip())
        total += number
        count += 1

# Calculate the average
if count > 0:
    average = total / count
    # Display the average
    print(f"The average of the numbers is: {average}")
else:
    print("The file is empty.")

# The file is automatically closed when the with block is exited
```

### I predict that this will identify and display all of the numbers in the number.txt and tell us teh total and count of them.

# Question 9: exception handling.
Modify the program that you wrote for Exercise 6 so it handles the following exceptions:
It should handle any IOERROR exceptions that are raised when the file is opened and data is read from it.
It should handle any ValueError exceptions that are raised when the items that are read from the file are converted to a number.

## Ill do thsi all in one because its a big line of code but I will explain it in the coding commets.

```python
# Define the filename
filename = "numbers.txt"

try:
    # Initialize variables to store the sum of numbers and the count of numbers
    total = 0
    count = 0

    # Open the file in read mode
    with open(filename, "r") as file:
        # Read all lines from the file
        lines = file.readlines()
        
        # Process each line
        for line in lines:
            try:
                # Convert the line to an integer and add it to the total
                number = int(line.strip())
                total += number
                count += 1
            except ValueError:
                # Handle the exception if the conversion fails
                print(f"Could not convert line to integer: {line.strip()}")

    # Calculate the average
    if count > 0:
        average = total / count
        # Display the average
        print(f"The average of the numbers is: {average}")
    else:
        print("The file is empty or contains no valid numbers.")

except FileNotFoundError:
    print(f"The file '{filename}' does not exist.")
except IOError:
    print(f"An error occurred while reading the file '{filename}'.")

# The file is automatically closed when the with block is exited
```

# Programming ex: Question 2: File Head Display
Write a program that asks the user for the name of a file. The program should display only the first five lines of the file’s contents. If the file contains less than five lines, it should display the file’s entire contents.

## Function display_first_five_lines:This function takes a filename as an argument.It attempts to open the file in read mode. etc... we have to set up the main and the exception for errors in the files. we also want to identify the entry for the main to enter and leave. lets do that below. 

```python
def display_first_five_lines(filename):
    try:
        # Open the file in read mode
        with open(filename, "r") as file:
            # Read the first five lines of the file
            lines = file.readlines()
            
            # Determine the number of lines to display
            num_lines_to_display = min(5, len(lines))
            
            # Display the appropriate number of lines
            for i in range(num_lines_to_display):
                print(lines[i].strip())

    except FileNotFoundError:
        print(f"The file '{filename}' does not exist.")
    except IOError:
        print(f"An error occurred while reading the file '{filename}'.")

def main():
    # Ask the user for the name of the file
    filename = input("Enter the name of the file: ")
    
    # Display the first five lines of the file
    display_first_five_lines(filename)

if __name__ == "__main__":
    main()
```
