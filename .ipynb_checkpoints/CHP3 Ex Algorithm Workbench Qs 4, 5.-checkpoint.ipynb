Write nested decision structures that perform the following: If AMOUNT_1 is greater than 10 and AMOUNT_2 is less than 100, display the greater of AMOUNT_1 and AMOUNT_2. 

### SO the first thing you want to do is identify the varibles that you will be using in this case Amount 1 and 2.

```python
Amount_1 = int(input("Enter amount 1: "))
Amount_2 = int(input("Enter amount 2: "))
```

### Once thats done you can now start with the If statments and else statments plus formatting. 

```python
if Amount_1 > 10:
    if Amount_2 < 100:
        if Amount_1 > Amount_2:
            print("The greater amount is:", Amount_1)
        else:
            print("The greater amount is:", Amount_2)
    else:
        print("Amount_2 is not less than 100.")
else:
    print("Amount_1 is not greater than 10.")
```

### The formatting worked now time to combine them to make the whole code. 
My prediction is that the code will work for the numbers that its supposed to have and the code will work. 

```python
Amount_1 = int(input("Enter amount 1: "))
Amount_2 = int(input("Enter amount 2: "))

if Amount_1 > 10:
    if Amount_2 < 100:
        if Amount_1 > Amount_2:
            print("The greater amount is:", Amount_1)
        else:
            print("The greater amount is:", Amount_2)
    else:
        print("Amount_2 is not less than 100.")
else:
    print("Amount_1 is not greater than 10.") 
```
