# We will start on the next question now 
Please define a function named super_duper that accepts a number and then returns one of the following values:

The function returns "super" if the number is evenly divisible by 3.
The function returns "duper" if the number is evenly divisible by 7.
The function returns "superduper" if the number is evenly divisible by 3 and 7.
The function returns the number it was originally passed if the number is not divisible by either 3 or 7.
Finally, please define a main function that loops through numbers 51 through 99 (inclusive) as inputs to the super_duper function and prints each of the results. Your instructor will call this main method to test your solution.



## To get started we have to identify a main and a super duper so lets do that below 

```python
def super_duper(number):
    # Check if the number is divisible by both 3 and 7
    if number % 3 == 0 and number % 7 == 0:
        return "superduper"
    # Check if the number is divisible by 3 only
    elif number % 3 == 0:
        return "super"
    # Check if the number is divisible by 7 only
    elif number % 7 == 0:
        return "duper"
    # Return the number itself if it's not divisible by 3 or 7
    else:
        return number
```

## Thats the super duper now time to identify teh main 

```python
def main():
    # Loop through numbers from 51 to 99 inclusive
    for num in range(51, 100):
        # Get the result from the super_duper function
        result = super_duper(num)
        # Print the result
        print(result)

# Call the main function to test the solution
if __name__ == "__main__":
    main()
```

## Now combine them 

```python
def super_duper(number):
    # Check if the number is divisible by both 3 and 7
    if number % 3 == 0 and number % 7 == 0:
        return "superduper"
    # Check if the number is divisible by 3 only
    elif number % 3 == 0:
        return "super"
    # Check if the number is divisible by 7 only
    elif number % 7 == 0:
        return "duper"
    # Return the number itself if it's not divisible by 3 or 7
    else:
        return number

def main():
    # Loop through numbers from 51 to 99 inclusive
    for num in range(51, 100):
        # Get the result from the super_duper function
        result = super_duper(num)
        # Print the result
        print(result)

# Call the main function to test the solution
if __name__ == "__main__":
    main()
```

# Finally the last question that We have here today 
Please create two classes that implement all of the names, data attributes, and methods reflected in the following UML diagram. These classes will be used within an inventory management system for ACME Eyewear.

The __str__() method in both classes should create a reasonable string representation for use in calls to the Python print() function.

```python
class Glasses:
    def __init__(self, id, name, brand, price, quantity):
        self.id = id
        self.name = name
        self.brand = brand
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Glasses(ID: {self.id}, Name: {self.name}, Brand: {self.brand}, Price: ${self.price:.2f}, Quantity: {self.quantity})"

    def update_quantity(self, quantity):
        self.quantity = quantity

    def get_total_values(self):
        return self.price * self.quantity

class Bifocals(Glasses):
    def __init__(self, id, name, brand, price, quantity, lens_type):
        super().__init__(id, name, brand, price, quantity)
        self.lens_type = lens_type
     
    def __str__(self):
        return f"Bifocals(ID: {self.id}, Name: {self.name}, Brand: {self.brand}, Price: ${self.price:.2f}, Quantity: {self.quantity}, Lens Type: {self.lens_type})"

    def get_total_value(self):
        return super().get_total_value()
```
