## Question 1: Total Sales:
Design a program that asks the user to enter a storeâ€™s sales for each day of the week. The amounts should be stored in a list. Use a loop to calculate the total sales for the week and display the result. 

## First you want to define the main and make a string of all teh days of the week. from there you go into asking teh user for all the sales input. Then calculate the total sales for the week then print it !  

```python
def main():
    # List to store sales amounts
    sales = []
    
    # Days of the week
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    
    # Ask the user to enter sales for each day
    for day in days_of_week:
        while True:
            try:
                amount = float(input(f"Enter sales for {day}: "))
                sales.append(amount)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

    # Calculate the total sales for the week
    total_sales = sum(sales)
    
    # Display the total sales
    print(f"\nTotal sales for the week: ${total_sales:.2f}")

if __name__ == "__main__":
    main()
```

Create a text file that contains your expenses for last month in the following categories:
 Rent

Write a Python program that reads the data from the file and uses matplotlib to plot a pie chart showing how you spend your money.

## First you want to import matplotlib and then you want to then go into teh files znd make a file for monthly expenses and then you want to identify the code to make a pie chart with the information

```python
expenses_content = """Rent 1200
Gas 150
Food 400
Clothing 100
Car payment 300
Misc 200
"""

# Define the filename
filename = "expenses.txt"
```

## THis idetifys the expenses in the code 

```python
import matplotlib.pyplot as plt

# Function to read expenses from the file and store them in a dictionary
def read_expenses(filename):
    expenses = {}
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            category = ' '.join(parts[:-1])
            amount = float(parts[-1])
            expenses[category] = amount
    return expenses

# Function to plot a pie chart
def plot_expenses(expenses):
    categories = list(expenses.keys())
    amounts = list(expenses.values())
    
    plt.figure(figsize=(8, 8))
    plt.pie(amounts, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.title("Monthly Expenses")
    plt.show()

def main():
    filename = 'expenses.txt'
    expenses = read_expenses(filename)
    plot_expenses(expenses)

if __name__ == "__main__":
    main()
```
