## Question 1: Total Sales:
Design a program that asks the user to enter a storeâ€™s sales for each day of the week. The amounts should be stored in a list. Use a loop to calculate the total sales for the week and display the result. 

## First you want to define the main and make a string of all teh days of the week. from there you go into asking teh user for all the sales input. Then calculate the total sales for the week then print it !  

```python
def main():
    # List to store sales amounts
    sales = []
    
    # Days of the week
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    
    # Ask the user to enter sales for each day
    for day in days_of_week:
        while True:
            try:
                amount = float(input(f"Enter sales for {day}: "))
                sales.append(amount)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

    # Calculate the total sales for the week
    total_sales = sum(sales)
    
    # Display the total sales
    print(f"\nTotal sales for the week: ${total_sales:.2f}")

if __name__ == "__main__":
    main()
```

## THis idetifys the expenses in the code 

```python
expenses_content = """Rent 1200
Gas 150
Food 400
Clothing 100
Car payment 300
Misc 200
"""

# Define the filename
filename = "expenses.txt"

# Write the content to the file
with open(filename, "w") as file:
    file.write(expenses_content)




import matplotlib.pyplot as plt

# Function to read expenses from the file and store them in a dictionary
def read_expenses(filename):
    expenses = {}
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            category = ' '.join(parts[:-1])
            amount = float(parts[-1])
            expenses[category] = amount
    return expenses

# Function to plot a pie chart
def plot_expenses(expenses):
    categories = list(expenses.keys())
    amounts = list(expenses.values())
    
    plt.figure(figsize=(8, 8))
    plt.pie(amounts, labels=categories, autopct='%1.1f%%', startangle=140)
    plt.title("Monthly Expenses")
    plt.show()

def main():
    filename = 'expenses.txt'
    expenses = read_expenses(filename)
    plot_expenses(expenses)

if __name__ == "__main__":
    main()
```

Design a program that generates a seven-digit lottery number. The program should generate seven random numbers, each in the range of 0 through 9, and assign each number to a list element. (Random numbers were discussed in Chapter 5.) Then write another loop that displays the contents of the list.

## first you want to import random and then make a file that is empty so it can take in random number files and print them out each time so lets do that below 

```python
import random

def generate_lottery_numbers():
    # Create an empty list to store the lottery numbers
    lottery_numbers = []
    
    # Generate seven random numbers, each between 0 and 9
    for _ in range(7):
        number = random.randint(0, 9)
        lottery_numbers.append(number)
    
    return lottery_numbers

def display_lottery_numbers(lottery_numbers):
    # Display the contents of the list
    for number in lottery_numbers:
        print(number, end="")
    print()  # For a newline at the end
```

# Question 3: Rainfall statistics.
Design a program that lets the user enter the total rainfall for each of 12 months into a list. The program should calculate and display the total rainfall for the year, the average monthly rainfall, the months with the highest and lowest amounts.

### From here you then want to call back to the main and make sure you have calculations that average the rainfall for the 12 months and tell you what the avrage is fort he rainfall. 

```python
def get_rainfall_data():
    # List to store rainfall amounts
    rainfall = []
    
    # List of month names for reference
    months = ["January", "February", "March", "April", "May", "June", 
              "July", "August", "September", "October", "November", "December"]
    
    # Get rainfall data from the user
    for month in months:
        while True:
            try:
                amount = float(input(f"Enter the total rainfall for {month}: "))
                if amount < 0:
                    print("Rainfall cannot be negative. Please enter a valid amount.")
                else:
                    rainfall.append(amount)
                    break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")
    
    return rainfall, months

def calculate_total_rainfall(rainfall):
    # Calculate the total rainfall for the year
    return sum(rainfall)

def calculate_average_rainfall(rainfall):
    # Calculate the average monthly rainfall
    return sum(rainfall) / len(rainfall)

def find_highest_rainfall(rainfall, months):
    # Find the month with the highest rainfall
    max_rainfall = max(rainfall)
    max_month = months[rainfall.index(max_rainfall)]
    return max_month, max_rainfall

def find_lowest_rainfall(rainfall, months):
    # Find the month with the lowest rainfall
    min_rainfall = min(rainfall)
    min_month = months[rainfall.index(min_rainfall)]
    return min_month, min_rainfall

def main():
    # Get rainfall data from the user
    rainfall, months = get_rainfall_data()
    
    # Calculate total and average rainfall
    total_rainfall = calculate_total_rainfall(rainfall)
    average_rainfall = calculate_average_rainfall(rainfall)
    
    # Find the months with the highest and lowest rainfall
    max_month, max_rainfall = find_highest_rainfall(rainfall, months)
    min_month, min_rainfall = find_lowest_rainfall(rainfall, months)
    
    # Display the results
    print("\nRainfall Data Analysis")
    print(f"Total rainfall for the year: {total_rainfall:.2f} inches")
    print(f"Average monthly rainfall: {average_rainfall:.2f} inches")
    print(f"Month with the highest rainfall: {max_month} ({max_rainfall:.2f} inches)")
    print(f"Month with the lowest rainfall: {min_month} ({min_rainfall:.2f} inches)")

if __name__ == "__main__":
    main()
```
