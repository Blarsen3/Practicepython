## Question 1: 
Write a statement that creates a list with the following strings: Einstein, Newton, Copernicus, and Kepler

### THis code is very straight forward as I am just identifying the strings and making them into a list as I will do below. 

```python
#I Identify the scientists varible and it has the strings inside of the statment. 
scientists = ["Einstein", "Newton", "Copernicus", "Kepler"]
```

## If we wnat to we can always recall to the list again.

```python
# Print the list of scientists
print(scientists)

# Access individual elements
print(scientists[0])  # Output: Einstein
print(scientists[1])  # Output: Newton
print(scientists[2])  # Output: Copernicus
print(scientists[3])  # Output: Kepler
```

# Question 2:
Assume names references a list. Write a for loop that displays each element of the list.

## The first step is to use the identify the names list and then to initiate the for and loop statmeant 

```python
names = ["Einstein", "Newton", "Copernicus", "Kepler"]  

# For loop to display each element in the list
for name in names:
    print(name)
```
