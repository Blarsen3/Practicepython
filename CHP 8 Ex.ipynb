## Question 10: Most frequent Character
Write a program that lets the user enter a string and displays the character that appears most frequently in the string.

### First we want to Create a dictionary to count the frequency of each character and then Find the character with the highest frequency and lets do that below 

```python
def most_frequent_character(input_string):
    # Create a dictionary to count the frequency of each character
    char_count = {}
    
    for char in input_string:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    
    # Find the character with the highest frequency
    most_frequent = max(char_count, key=char_count.get)
    
    return most_frequent, char_count[most_frequent]
```
