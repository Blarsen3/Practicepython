### Next we will want to Get the initials and then display the initials so we will do that in the code below and see if that works.

```python
# Get the user's full name
full_name = input("Enter your first, middle, and last names: ")

# Split the name into parts
name_parts = full_name.split()

# Get the initials
initials = ''
for name in name_parts:
    initials += name[0].upper() + '. '

# Display the initials
print("Your initials are:", initials.strip())
```

### Now we want to split the date and then Get the month and the day and then print the month day and year in the correct format lets do that below. 

```python
# Define a dictionary to map month numbers to month names
month_names = {
    "01": "January", "02": "February", "03": "March", "04": "April",
    "05": "May", "06": "June", "07": "July", "08": "August",
    "09": "September", "10": "October", "11": "November", "12": "December"
}

# Get the date from the user
date_string = input("Enter a date in the form mm/dd/yyyy: ")

# Split the date into month, day, and year
month, day, year = date_string.split('/')

# Get the month name from the dictionary
month_name = month_names[month]

# Print the formatted date
print(f"{month_name} {int(day)}, {year}")
```
