## Question 2: 
Write a loop that counts the number of space characters that appear in the string referenced by mystring

### First you want to initiate the my string varible and set the space count to begin the code shown below.

```python
mystring = "Your example string here with spaces."
space_count = 0

for char in mystring:
    if char == ' ':
        space_count += 1
```
