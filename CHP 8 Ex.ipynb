## Question 2: 
Write a loop that counts the number of space characters that appear in the string referenced by mystring

### First you want to initiate the my string varible and set the space count to begin the code shown below.

```python
mystring = "Your example string here with spaces."
space_count = 0

for char in mystring:
    if char == ' ':
        space_count += 1
```

### I then added the space count to 1 so it counts every single one.

### Now then you want to add the print statment of the my string and add the space count into it. 


```python
mystring = "Your example string here with spaces."
space_count = 0
# Adding the my string and the space count to go up. 
for char in mystring:
    if char == ' ':
        space_count += 1
# Print the statment with the space count also included. 
print("Number of space characters:", space_count)
```
