Write a program that asks the user for the speed of a vehicle (in miles per hour) and the number of hours it has traveled. It should then use a loop to display the distance the vehicle has traveled for each hour of that time period. |

## Start by writing the print statments that are the questions it will ask and then also initalize any variables you have.

## From there you will want to write the equation and any other small things you will need. 

```python
speed = float(input("Enter the speed of the vehicle (in miles per hour): "))

hours = int(input("Enter the number of hours the vehicle has traveled: "))

print("\nHour\tDistance Traveled")
print("-------------------------")

for hour in range(1, hours + 1):

    distance = speed * hour
    
    print(f"{hour}\t{distance:.2f}")
```
