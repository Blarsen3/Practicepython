## From there you will want to write the equation and any other small things you will need. 

```python
speed = float(input("Enter the speed of the vehicle (in miles per hour): "))

hours = int(input("Enter the number of hours the vehicle has traveled: "))

print("\nHour\tDistance Traveled")
print("-------------------------")

for hour in range(1, hours + 1):

    distance = speed * hour
    
    print(f"{hour}\t{distance:.2f}")
```

## Next you wnat to write a for statment and any other small print statments to enclose the boundry of this problem

```python
starting_weight = float(input("Enter your starting weight (in pounds): "))

weight_loss_per_month = 4

print("\nMonth\tExpected Weight")
print("-------------------------")

for month in range(1, 7):

    expected_weight = starting_weight - (weight_loss_per_month * month)
    

    print(f"{month}\t{expected_weight:.2f}")

```
