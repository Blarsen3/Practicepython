## Question 1: Total Sales:
Design a program that asks the user to enter a storeâ€™s sales for each day of the week. The amounts should be stored in a list. Use a loop to calculate the total sales for the week and display the result. 

## First you want to define the main and make a string of all teh days of the week. from there you go into asking teh user for all the sales input. Then calculate the total sales for the week then print it !  

```python
def main():
    # List to store sales amounts
    sales = []
    
    # Days of the week
    days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    
    # Ask the user to enter sales for each day
    for day in days_of_week:
        while True:
            try:
                amount = float(input(f"Enter sales for {day}: "))
                sales.append(amount)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

    # Calculate the total sales for the week
    total_sales = sum(sales)
    
    # Display the total sales
    print(f"\nTotal sales for the week: ${total_sales:.2f}")

if __name__ == "__main__":
    main()
```
