## Programming Ex Question 1
Write a program that displays the following information:
                
Your name, Your address, with city, state, and ZIP, Your telephone number, Your college major

### First you want to identify the variables.

```python
name = "Bobby Larsen"
address = "123 Main St, Springfield, IL, 62701"
telephone_number = "555-123-4567"
college_major = "Business Marketing" 
```

### After Identifying Variables Display the information in a way thats a list in order:

```python
print("Name:", name)
print("Address:", address)
print("Telephone Number:", telephone_number)
print("College Major:", college_major) 
```


### Once you have that verified you combine them:

```python
name = "Bobby Larsen"
address = "123 Main St, Springfield, IL, 62701"
telephone_number = "555-123-4567"
college_major = "Business Marketing"

print("Name:", name)
print("Address:", address)
print("Telephone Number:", telephone_number)
print("College Major:", college_major) 
```

## Boom thats how you write a program to display your information.
