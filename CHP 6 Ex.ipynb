# Question 6: Average of numbers.
Assume a file containing a series of integers is named numbers.txt and exists on the computer’s disk. Write a program that calculates the average of all the numbers stored in the file

## First you want to Define the filename: The file containing the integers is named numbers.txt. Initialize variables: Variables total and count are initialized to store the sum of the numbers and the count of the numbers, respectively.Open the file in read mode: The open function with the with statement opens the file in read mode ("r").Read all lines: The readlines method reads all lines from the file into a list.

Lest do that below 

```python
# Define the filename
filename = "numbers.txt"

# Open the file in write mode
with open(filename, "w") as file:
    # Write your name to the file
    file.write("25 ")



# Define the filename
filename = "numbers.txt"

# Initialize variables to store the sum of numbers and the count of numbers
total = 0
count = 0

# Open the file in read mode
with open(filename, "r") as file:
    # Read all lines from the file
    lines = file.readlines()
    
    # Process each line
    for line in lines:
        # Convert the line to an integer and add it to the total
        number = int(line.strip())
        total += number
        count += 1

# Calculate the average
if count > 0:
    average = total / count
    # Display the average
    print(f"The average of the numbers is: {average}")
else:
    print("The file is empty.")

# The file is automatically closed when the with block is exited
```
