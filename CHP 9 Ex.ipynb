## FIrst we want to start by writing a file input and output value and on top of that also write a dictonary and read the input file and then use the whole file.

```python
def count_word_frequencies(input_file, output_file=None): 
    # Dictionary to store word frequencies
    word_freq = {}
    # Read the input file
    with open(input_file, 'r') as file:
        # Read entire content of the file
        content = file.read()
        # split the content into words
        words = content.split()
        # Count frequencies of each word 
        for word in words:
            # remove punctuation from the word (if any)
            word = word.strip('.,!?-').lower()
            # Update the dictonary
            if word:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
```
