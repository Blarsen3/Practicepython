## Question 1:

Write a program that asks the user for a number in the range of 1 through 7. The program should display the corresponding day of the week, where 1 = Monday, 2 = Tuesday, 3 = Wednesday, 4 = Thursday, 5 = Friday, 6 = Saturday, and 7 = Sunday. The program should display an error message if the user enters a number that is outside the range of 1 through 7.

### So the first thing you should start with is identifying the input line of code so they can put a number in. 

```python
number = int(input("Enter a number between 1 and 7: "))
```

### This allows us to input any number that is needed for this specific problem. 
Next we want to begin by making the IF and ElSE statments that are needed in this problem. With that being said we have to identigy the numbers of days and Correlate them to days of the week. 

```python
if 1 <= number <= 7:
    if number == 1:
        day = "Monday"
    elif number == 2:
        day = "Tuesday"
    elif number == 3:
        day = "Wednesday"
    elif number == 4:
        day = "Thursday"
    elif number == 5:
        day = "Friday"
    elif number == 6:
        day = "Saturday"
    else:
        day = "Sunday"
```

## Finally you want to add the print statments in the bottom underneath all the if statments anc combine them.
My prediction is that the code will ask me for a number and if I put a number that is 1-7 It will give me a response and work and if I enter a Number higher then 7 It will tell me to enter a number in between 1-7.

```python
number = int(input("Enter a number between 1 and 7: "))

if 1 <= number <= 7:

    if number == 1:
        day = "Monday"
    elif number == 2:
        day = "Tuesday"
    elif number == 3:
        day = "Wednesday"
    elif number == 4:
        day = "Thursday"
    elif number == 5:
        day = "Friday"
    elif number == 6:
        day = "Saturday"
    else:
        day = "Sunday"
    
 
    print("The corresponding day of the week is:", day)
else:
   
    print("Error: Please enter a number between 1 and 7.")
```
