```python
if weight > 500:
    print("The object is too heavy.")
elif weight < 100:
    print("The object is too light.")
else:
    print("The weight of the object is:", weight, "newtons.")
```

# Question 16:
The month of February normally has 28 days. But if it is a leap year, February has 29 days. Write a program that asks the user to enter a year. The program should then display the number of days in February that year. Use the following criteria to identify leap years:
                
Determine whether the year is divisible by 100. If it is, then it is a leap year if and only if it is also divisible by 400. For example, 2000 is a leap year, but 2100 is not.
                
If the year is not divisible by 100, then it is a leap year if and only if it is divisible by 4. For example, 2008 is a leap year, but 2009 is not.

### First off they are asking us to have a program where the user, Enters a year, displays the number of days in february that year.
First things first Ill start with the simple stuff and identify the input line of code. 

```python
year = int(input("Enter a year: "))
```

### From there now we have to start writing teh code for the equations that need to be done. 

```python
year = int(input("Enter a year: "))

if (year % 100 == 0 and year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
    days_in_february = 29
else:
    days_in_february = 28
```

### and finally we finish this one off with just a little print statement and the code shuld work !!

```python
year = int(input("Enter a year: "))

if (year % 100 == 0 and year % 400 == 0) or (year % 100 != 0 and year % 4 == 0):
    days_in_february = 29
else:
    days_in_february = 28

print(f"In {year}, February has {days_in_february} days.")
```
