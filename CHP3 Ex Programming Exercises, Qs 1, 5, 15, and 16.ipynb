# Question 5:
Scientists measure an object’s mass in kilograms and its weight in newtons. If you know the amount of mass of an object in kilograms, you can calculate its weight in newtons with the following formula:

Weight = mass × 9.8

Write a program that asks the user to enter an object’s mass, then calculates its weight. If the object weighs more than 500 newtons, display a message indicating that it is too heavy. If the object weighs less than 100 newtons, display a message indicating that it is too light.

## The first step is to identify the Mass and the weight because they need a value to enter and go into the equation and Identify the Equation.  

```python
mass = float(input("Enter the mass of the object in kilograms: "))

weight = mass * 9.8
```

### Once you get the simple part done with you now have to go and write the print, elif, and else statments. 

```python
if weight > 500:
    print("The object is too heavy.")
elif weight < 100:
    print("The object is too light.")
else:
    print("The weight of the object is:", weight, "newtons.")
```

### Now combine them and see the final result and enter values to see if it worked or not. 
My prediction is that this equation will work the way it is supposed to and will print the weight needed. 

```python
mass = float(input("Enter the mass of the object in kilograms: "))

weight = mass * 9.8

if weight > 500:
    print("The object is too heavy.")
elif weight < 100:
    print("The object is too light.")
else:
    print("The weight of the object is:", weight, "newtons.")
```

## As you can see the code works and tells you the weight of an object in kilograms and tells you the weight in newtons.
